```python
# prompt: Write a program to perform element-wise multiplication of two NumPy array

import numpy as np

def element_wise_multiplication(array1, array2):
  """
  Performs element-wise multiplication of two NumPy arrays.

  Args:
    array1: The first NumPy array.
    array2: The second NumPy array.

  Returns:
    A new NumPy array representing the element-wise product, or None if the arrays are not compatible.
  """
  if array1.shape != array2.shape:
    print("Error: Arrays must have the same shape for element-wise multiplication.")
    return None

  return array1 * array2

# Example usage
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
result_array = element_wise_multiplication(arr1, arr2)

if result_array is not None:
  print("Element-wise multiplication result:", result_array)


arr3 = np.array([[1, 2], [3, 4]])
arr4 = np.array([[5, 6], [7, 8]])
result_array = element_wise_multiplication(arr3, arr4)

if result_array is not None:
  print("Element-wise multiplication result:", result_array)
```

```python
# prompt: Perform matrix multiplication using NumPy

import numpy as np

def matrix_multiply(matrix1, matrix2):
    """
    Performs matrix multiplication of two matrices.

    Args:
        matrix1: The first matrix (NumPy array).
        matrix2: The second matrix (NumPy array).

    Returns:
        The resulting matrix product (NumPy array), or None if the matrices
        are not compatible for multiplication.
    """
    try:
      result = np.dot(matrix1, matrix2)
      return result
    except ValueError:
      print("Matrices are not compatible for multiplication.")
      return None

# Example usage
matrix_a = np.array([[1, 2], [3, 4]])
matrix_b = np.array([[5, 6], [7, 8]])

product_matrix = matrix_multiply(matrix_a, matrix_b)

if product_matrix is not None:
    print("Matrix Product:")
product_matrix
```
